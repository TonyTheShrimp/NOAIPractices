### CSV 文件格式
逗号分隔值（Comma-Separated Values，CSV），其文件以纯文本形式存储表格数据（数字和文本），文件的每一行都是一个数据记录。  
每个记录由一个或多个字段组成，用逗号分隔。使用逗号作为字段分隔符是此文件格式的名称的来源，因为分隔字符也可以不是逗号，有时也称为字符分隔值。  
可以简单把CSV文件理解问一张文本化的Excel表格，因为是文本文件格式的，所以CSV文件便于在不同系统之间分享数据。也可以被各种语言处理。

#### CSV文件标准如下：  
<b>1. 每一行记录位于一个单独的行上，用回车换行符CRLF(也就是\r\n)分割。</b>  
aaa,bbb,ccc <font color="red">CRLF</font>  
zzz,yyy,xxx <font color="red">CRLF</font>  

<b>2. 文件中的最后一行记录可以有结尾回车换行符，也可以没有。</b>  
aaa,bbb,ccc <font color="red">CRLF</font>  
zzz,yyy,xxx <font color="red">*CRLF*</font>  

<b>3. 第一行可以存在一个可选的标题头，格式和普通记录行的格式一样。标题头要包含文件记录字段对应的名称，应该有和记录字段一样的数量。</b>  
field_name,field_name,field_name <font color="red">CRLF</font>
aaa,bbb,ccc <font color="red">CRLF</font>
zzz,yyy,xxx <font color="red">CRLF</font> 

<b>4. 在标题头行和普通行每行记录中，会存在一个或多个由半角逗号(,)分隔的字段。整个文件中每行应包含相同数量的字段，空格也是字段的一部分，不应被忽略。每一行记录最后一个字段后不能跟逗号。（通常用逗号分隔，也有其他字符分隔的CSV，需事先约定）</b>  
aaa,bbb,ccc

<b>5. 每个字段可用也可不用半角双引号(“)括起来（不过有些程序，如Microsoft的Excel就根本不用双引号）。如果字段没有用引号括起来，那么该字段内部不能出现双引号字符。</b>  
aaa,bbb,ccc <font color="red">CRLF</font>   
zzz,yyy,xxx <font color="red">CRLF</font>   

<b>6. 字段中若包含回车换行符、双引号或者逗号，该字段需要用双引号括起来。</b>  
"aaa","b CRLF
bb","ccc " <font color="red">CRLF</font>  
zzz,yyy,xxx

<b>7. 如果用双引号括字段，那么出现在字段内的双引号前必须加一个双引号进行转义。两个""表示一个“。</b>  
"aaa","b""bb","ccc"
