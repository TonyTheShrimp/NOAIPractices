```python
# !pip install brewer2mpl
import numpy as np
from random import randint
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')



large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_style("white")
%matplotlib inline

# 设置中文显示
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体

# Version
print(mpl.__version__)  #> 3.0.0
print(sns.__version__)  #> 0.9.0
```

### 图表大小
在 matplotlib 中，设置图表（Figure）的大小是通过 figure() 函数中的 figsize 参数来控制的。figsize 参数接受一个元组，表示图表的宽度和高度（单位为英寸）。

### 横纵轴上面数字的大小  
可以修改横纵轴上面数字的大小
plt.xticks(fontsize=12); plt.yticks(fontsize=12)

### 图表角上标注文字的字体大小,loc表示位置
plt.legend(fontsize=10, loc='upper left',frameon=True, title='Legend',shadow=True)  
**可以选的位置：**  
'best' (0): 自动选择最不碍碍看的位置（默认值）。选择逻辑是尽量不遮挡数据。  
'upper right' (1): 将图例放在图的右上角。  
'upper left' (2): 将图例放在图的左上角。  
'lower left' (3): 将图例放在图的左下角。  
'lower right' (4): 将图例放在图的右下角。  
'right' (5): 将图例放在图的右侧。  
'center left' (6): 将图例放在图的中左。  
'center right' (7): 将图例放在图的中右。  
'lower center' (8): 将图例放在图的底部中央。  
'upper center' (9): 将图例放在图的顶部中央。  
'center' (10): 将图例放在图的中心。  


```python
# 设置图表大小为 3x2 英寸
plt.figure(figsize=(3, 3))

# 生成一些数据并绘图
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.plot(x, y, label="Line 1")

# 设置标题和坐标轴标签
plt.title('我是一张图表')
plt.xlabel('我是X轴')
plt.ylabel('我是Y轴')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=10, loc='upper left',frameon=True, title='Legend',shadow=True)  

# 显示图表
plt.show()

# 设置图表大小为 6x4 英寸
plt.figure(figsize=(6, 4))

# 生成一些数据并绘图
x = [5, 4, 3, 2, 1]
y = [2, 3, 4, 5, 6]
plt.plot(x, y, label="Line 02")

# 设置标题和坐标轴标签
plt.title('我是第二张图表')
plt.xlabel('我是X轴')
plt.ylabel('我是Y轴')
plt.xticks(fontsize=8)
plt.yticks(fontsize=14)
plt.legend(fontsize=10, loc='upper right',frameon=True, title='Legend',shadow=True)  

# 显示图表
plt.show()

# 设置图表大小为 8x4 英寸
plt.figure(figsize=(8, 4))

# 生成一些数据并绘图
x = [5, 3, 1, 4, 5]
y = [2, 3, 4, 5, 6]
plt.plot(x, y)

# 设置标题和坐标轴标签
plt.title('我是第三张图表')
plt.xlabel('我是X轴')
plt.ylabel('我是Y轴')
plt.xticks(fontsize=15)
plt.yticks(fontsize=10)

# 显示图表
plt.show()
```

### 多个图表并列布局

在 matplotlib 中，如果你想在一个页面上布置多个图表，可以使用几种方法，如 subplot、subplots，或者更高级的布局管理工具如 GridSpec。这些工具提供了灵活的方式来安排多个图表。  

#### 使用 subplot 方法
subplot 方法允许你在一个 figure 对象中创建多个子图。每个子图都可以独立绘制不同的内容。


```python
# 创建一个 figure
plt.figure(figsize=(10, 5))

# 第一个子图
# 子图占一行，每行两个子图，下面画第一个子图（从1开始编号，先行后列）
plt.subplot(2, 3, 1)  # (行数, 列数, 子图编号)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('第一个子图')

# 第二个子图
plt.subplot(2, 3, 2)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('第二个子图')

# 第三个子图
plt.subplot(2, 3, 4)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('第三个子图')

# 第四个子图
plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('第四个子图')

# 自动调整布局，没有这句的话，Title和图可能重叠
plt.tight_layout()

# 显示图表
plt.show()
```

#### 使用 GridSpec 方法
对于更复杂的子图布局，GridSpec 提供了更高级的布局选项。例如，你可以使某些子图跨越多个行或列。

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 8))
gs = GridSpec(3, 3, figure=fig)

ax1 = fig.add_subplot(gs[0, :])
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax1.set_title('Subplot 1: Span all columns')

ax2 = fig.add_subplot(gs[1, :-1])
ax2.plot([1, 2, 3, 4], [1, 2, 3, 4])
ax2.set_title('Subplot 2: Span first two columns')

ax3 = fig.add_subplot(gs[1:, -1])
ax3.plot([1, 2, 3, 4], [4, 3, 2, 1])
ax3.set_title('Subplot 3: Span last two rows')

ax4 = fig.add_subplot(gs[2, 0])
ax4.plot([1, 2, 3, 4], [16, 9, 4, 1])
ax4.set_title('Subplot 4: Bottom-left corner')

plt.tight_layout()
plt.show()
```

### matplotlib也有预定义的颜色表格
在 matplotlib 中，有多种颜色映射表（colormap），这些颜色映射表通常分为几个类别：连续（Sequential）、离散（Qualitative）、和发散（Diverging）颜色映射。这些颜色映射表广泛应用于各种数据可视化任务，以帮助解释和突出显示数据特征。　　

##### 1. 连续颜色映射（Sequential colormaps）　　
这些颜色映射通常用于表示从低到高的数值变化，其颜色从浅到深，适用于表示数量的增减。一些常用的连续颜色映射包括：　　
viridis　　
plasma　　
inferno　　
magma　　
cividis　　

##### 2. 离散颜色映射（Qualitative colormaps）  
离散颜色映射提供明显区分的颜色，适合用来表示没有顺序或数值关系的类别数据。除了 tab10，还有：  
tab10  
tab20  
tab20b  
tab20c  
Pastel1  
Pastel2  
Set1  
Set2  
Set3  
Accent  
##### 3. 发散颜色映射（Diverging colormaps）
发散颜色映射通常用于显示数据从中间基准值向两边变化的情况，颜色在中间具有共同的起始点，向两端变化成对比色。这些映射非常适合于突出中间区域的偏差或极值。常用的发散颜色映射包括：  
coolwarm  
bwr  
seismic  
PiYG  
PRGn  
RdBu  
RdYlBu  
RdYlGn  
Spectral  

```python
# 预定义颜色表
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# 创建数据
data = np.random.randn(10, 10)

# 使用不同的颜色映射绘制图像
plt.figure(figsize=(12, 8))

plt.subplot(441)
plt.title('Sequential: viridis')
plt.imshow(data, cmap='viridis')
plt.colorbar()

plt.subplot(442)
plt.title('Qualitative: plasma')
plt.imshow(data, cmap='plasma')
plt.colorbar()

plt.subplot(443)
plt.title('Diverging: inferno')
plt.imshow(data, cmap='inferno')
plt.colorbar()

plt.subplot(444)
plt.title('Diverging: magma')
plt.imshow(data, cmap='magma')
plt.colorbar()

plt.subplot(445)
plt.title('Diverging: cividis')
plt.imshow(data, cmap='cividis')
plt.colorbar()

plt.subplot(447)
plt.title('Diverging: tab10')
plt.imshow(data, cmap='tab10')
plt.colorbar()

plt.subplot(448)
plt.title('Diverging: tab20')
plt.imshow(data, cmap='tab20')
plt.colorbar()

plt.subplot(449)
plt.title('Diverging: tab20b')
plt.imshow(data, cmap='tab20b')
plt.colorbar()

plt.subplot(4,4,10)
plt.title('Diverging: tab20c')
plt.imshow(data, cmap='tab20c')
plt.colorbar()

plt.subplot(4,4,11)
plt.title('Diverging: Pastel1')
plt.imshow(data, cmap='Pastel1')
plt.colorbar()

plt.subplot(4,4,12)
plt.title('Diverging: Pastel2')
plt.imshow(data, cmap='Pastel2')
plt.colorbar()

plt.subplot(4,4,13)
plt.title('Diverging: Set1')
plt.imshow(data, cmap='Set1')
plt.colorbar()

plt.subplot(4,4,14)
plt.title('Diverging: Set2')
plt.imshow(data, cmap='Set2')
plt.colorbar()

plt.subplot(4,4,15)
plt.title('Diverging: Set3')
plt.imshow(data, cmap='Set3')
plt.colorbar()

plt.subplot(4,4,16)
plt.title('Diverging: Accent')
plt.imshow(data, cmap='Accent')
plt.colorbar()

plt.tight_layout()
plt.show()

# 使用不同的颜色映射绘制图像
plt.figure(figsize=(12, 8))

plt.subplot(331)
plt.title('Sequential: coolwarm')
plt.imshow(data, cmap='coolwarm')
plt.colorbar()

plt.subplot(332)
plt.title('Sequential: bwr')
plt.imshow(data, cmap='bwr')
plt.colorbar()

plt.subplot(333)
plt.title('Sequential: seismic')
plt.imshow(data, cmap='seismic')
plt.colorbar()

plt.subplot(334)
plt.title('Sequential: PiYG')
plt.imshow(data, cmap='PiYG')
plt.colorbar()

plt.subplot(335)
plt.title('Sequential: PRGn')
plt.imshow(data, cmap='PRGn')
plt.colorbar()

plt.subplot(336)
plt.title('Sequential: RdBu')
plt.imshow(data, cmap='RdBu')
plt.colorbar()

plt.subplot(337)
plt.title('Sequential: RdYlBu')
plt.imshow(data, cmap='RdYlBu')
plt.colorbar()

plt.subplot(338)
plt.title('Sequential: RdYlGn')
plt.imshow(data, cmap='RdYlGn')
plt.colorbar()

plt.subplot(339)
plt.title('Sequential: Spectral')
plt.imshow(data, cmap='Spectral')
plt.colorbar()



plt.tight_layout()
plt.show()

```
