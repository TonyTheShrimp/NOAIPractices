鸢尾花数据也是机器学习的一个著名的数据集。  
Iris鸢尾花数据集是一个经典的数据集，由统计学家和生物学家Ronald A. Fisher于1936年首次介绍。该数据集包含了三个不同种类的鸢尾花：山鸢尾（setosa）、变色鸢尾（versicolor）和维吉尼亚鸢尾（virginica）。每个类别包含50个样本，总共150个样本。每个样本都有四个特征：花萼长度（sepal length）、花萼宽度（sepal width）、花瓣长度（petal length）和花瓣宽度（petal width）。  

Iris鸢尾花数据集通常用于分类问题，即根据这四个特征来预测鸢尾花的种类。这是一个非常适合入门机器学习的数据集，因为它小巧且易于理解。  
  
sklearn的dataset模块自带了iris数据集。



## 1. 二维数据集--knn分类

**Step1: 库函数导入**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
```

**Step2: 数据导入**

```python
# 使用鸢尾花数据集的前两维数据，便于数据可视化
iris = datasets.load_iris()    # 加载自带的iris数据集
X = iris.data[:, :2]
y = iris.target
print("Data Number:"+str(len(iris.target)))
```

**Step3: 模型训练&可视化**

```python
k_list = [1, 3, 5, 8, 10, 15]
h = .1
# 创建不同颜色的画布
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

print("计算中。。。。等一下。。。。。。")

plt.figure(figsize=(15,14))
# 根据不同的k值进行可视化
for ind,k in enumerate(k_list):
    clf = KNeighborsClassifier(k)
    clf.fit(X, y)
    # 画出决策边界
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    # xx，yy是从_min到_max，以h为步长生成的索引一维数组
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    # Z实际是一个二维数组，保存的是二维平面上每一点用KNN算法预测的类别
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    # 根据类别填充颜色
    Z = Z.reshape(xx.shape)

    plt.subplot(321+ind)  
    # 这个绘图函数实际是拿xx，yy作为二维数组的X，Y索引，到Z这个二维数组中去取颜色显示。
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
    # 数据点可视化到画布
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i)"% k)

plt.show()
```

**Step4: 原理简析**

如果选择较小的K值，就相当于用较小的领域中的训练实例进行预测。
例如当k=1的时候，在分界点位置的数据很容易受到局部的影响，图中蓝色的部分中还有部分绿色块，主要是数据太局部敏感。
当k=15的时候，不同的数据基本根据颜色分开，当时进行预测的时候，会直接落到对应的区域，模型相对更加鲁棒。
