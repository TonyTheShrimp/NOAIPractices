### torchvision库简介
torchvision是pytorch的一个图形库，它服务于PyTorch深度学习框架的，主要用来构建计算机视觉模型。torchvision.transforms主要是用于常见的一些图形变换。

torchvision的构成如下：
torchvision.datasets: 一些加载数据的函数及常用的数据集接口；
torchvision.models: 包含常用的模型结构（含预训练模型），例如AlexNet、VGG、ResNet等；
torchvision.transforms: 常用的图片变换，例如裁剪、旋转等；

我们这里主要学习torchivsion的transforms能够对于图像进行的操作。

## 一、 裁剪——Crop
#### 1. 中心裁剪：transforms.CenterCrop 
**class torchvision.transforms.CenterCrop(size)**  
**功能：**依据给定的size从中心裁剪  
**参数：**  
size- (sequence or int)，若为sequence,则为([h,w])，若为int，则(size)  

注意参数是先高后宽，如果只有一个int作为size，则是正方形

```python
# 取得图片文件路径（join函数可以拼接每一层文件夹）
image_path = os.path.join( current_dir,"机器学习","PyTorch_tutorial_0.0.5_余霆嵩","panda96.png")

# padding_img = transforms.Pad(padding=10, fill=0)
# 加载图片文件
img96 = Image.open(image_path)

print("原图大小：",img96.size)
img01 = transforms.CenterCrop([64,32])(img96)
img02 = transforms.CenterCrop([32,64])(img96)
img03 = transforms.CenterCrop(32)(img96)
img04 = transforms.CenterCrop(64)(img96)

# plt subslot控制按格子显示，前两个参数表示3行两列，第三个参数表示第几个位置，从1开始变换，从左到右，从上到下。
plt.subplot(3,2,1)
plt.imshow(img96)
plt.title("原图")
plt.subplot(3,2,2)
plt.imshow(img01)
plt.title("转换后的图1")
plt.subplot(3,2,3)
plt.imshow(img02)
plt.title("转换后的图2")
plt.subplot(3,2,4)
plt.imshow(img03)
plt.title("转换后的图3")
plt.subplot(3,2,5)
plt.imshow(img04)
plt.title("转换后的图4")
plt.show()
```

#### 2. 随机裁剪：transforms.RandomCrop  
**class torchvision.transforms.RandomCrop(size)**  
**功能：**依据给定的size从随机位置为中心裁剪  
**参数：**  
size- (sequence or int)，若为sequence,则为([h,w])，若为int，则(size)  

注意参数是先高后宽，如果只有一个int作为size，则是正方形。
多运行下面的代码几次，可以看到每次取到的都不一样。

```python
print("原图大小：",img96.size)
img01 = transforms.RandomCrop([64,32])(img96)
img02 = transforms.RandomCrop([32,64])(img96)
img03 = transforms.RandomCrop(32)(img96)
img04 = transforms.RandomCrop(64)(img96)

# plt subslot控制按格子显示，前两个参数表示3行两列，第三个参数表示第几个位置，从1开始变换，从左到右，从上到下。

# plt.subplot2grid可以大致控制一下图片的位置和大小，但是好像并不太精确
# 而且有时候发生意外的重叠，有些内容就不能显示。
plt.figure()  # 创建图像
ax1 = plt.subplot2grid((4, 4),  # 三行三列
                       (0, 0),  # 从第0行第0列开始作图
                       colspan=2,  # 跨度为2
                       rowspan=2
                       )
ax1.imshow(img96) # 画小图
ax1.set_title('原图')  # 设置小图的标题
ax2 = plt.subplot2grid((4, 4), (0, 3), colspan=4)
ax2.imshow(img01) # 画小图
ax3 = plt.subplot2grid((4, 4), (2, 1), rowspan=2)
ax3.imshow(img02) # 画小图
ax4 = plt.subplot2grid((4, 4), (3, 0))
ax4.imshow(img03) # 画小图
ax5 = plt.subplot2grid((4, 4), (2, 2), rowspan=2, colspan=2)
ax5.imshow(img04) # 画小图

plt.show()
```

#### 3. 随机长宽比裁剪 transforms.RandomResizedCrop 
**class torchvision.transforms.CenterCrop(size, scale=(0.08, 1.0), ratio=(0.75, 1.3333333333333333), interpolation=2)**
**功能：**随机大小，随机长宽比裁剪原始图片，最后将图片resize到设定好的size  
**参数：**  
size- 输出的分辨率  
scale- 随机crop的大小区间，如scale=(0.08, 1.0)，表示随机crop出来的图片会在的0.08倍至1倍之间。  
ratio- 随机长宽比设置  
interpolation- 插值的方法，默认为双线性插值(PIL.Image.BILINEAR)  

同样，每次运行后得到的结果都不一样。

这个函数的用处主要在测试集，任意切分图片送进去测试。
也有RandomCrop，只随机取不自动Resize的。但是好像送进去测试都需要Resize。

```python
print("原图大小：",img96.size)
img01 = transforms.RandomResizedCrop([64,32],scale=(0.1,0.3),ratio=(0.75,1.33333))(img96)
img02 = transforms.RandomResizedCrop([32,64],scale=(2,4),ratio=(0.75,1.33333))(img96)
img03 = transforms.RandomResizedCrop(64,scale=(0.1,0.4),ratio=(0.75,1.33333))(img96)
img04 = transforms.RandomResizedCrop(64,scale=(0.6,0.8),ratio=(0.75,1.33333))(img96)

# plt subslot控制按格子显示，前两个参数表示3行两列，第三个参数表示第几个位置，从1开始变换，从左到右，从上到下。
plt.subplot(3,2,1)
plt.imshow(img96)
plt.title("原图")
plt.subplot(3,2,2)
plt.imshow(img01)
plt.title("转换后的图1")
plt.subplot(3,2,3)
plt.imshow(img02)
plt.title("转换后的图2")
plt.subplot(3,2,4)
plt.imshow(img03)
plt.title("转换后的图3")
plt.subplot(3,2,5)
plt.imshow(img04)
plt.title("转换后的图4")
plt.show()
```

#### 11. 标准化：transforms.Normalize
**class torchvision.transforms.Normalize(mean, std)** 
**功能：**  对数据按通道进行标准化，即先减均值，再除以标准差。经过这样处理后将数据标准化，即均值为0，标准差为1。使模型更容易收敛。注意data是 hwc
**参数：**

在图像处理中，HWC 是指图像的数据维度排列顺序：高度（Height）、宽度（Width）、和通道数（Channel）。这是一种常见的图像数据表示方式，尤其是在使用 Python 的图像处理库如 PIL 或者 OpenCV 时。
通常，你需要将图像从 HWC 转换为 CHW 格式来进行进一步处理。在 PyTorch 中，这可以通过 transforms.ToTensor() 完成，这个转换不仅将 PIL 图像或者 numpy 数组转换为张量，而且还会从 HWC 格式转换为 CHW 格式。

这里的mean和std参数，实际是应该从数据中计算来的。但是很多地方用的是mean=[0.485, 0.456, 0.406] 和 std=[0.229, 0.224, 0.225]。这些值是基于 ImageNet 数据集的大量图像计算得出的。ImageNet 是一个广泛用于计算机视觉研究和模型训练的大规模数据集，包括超过一百万张有注释的图像。

ToTensor和normalize都是为了方便后面的数学计算。

```python
normal = transforms.Normalize([0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])(tensor)
print(normal)
```

#### 12. 填充：transforms.Pad
**class torchvision.transforms.Pad(padding, fill=0, padding_mode=‘constant’)**   
**功能：**  对图像进行填充  
**参数：**  
padding-(sequence or int, optional)，此参数是设置填充多少个pixel。  
&nbsp;&nbsp;&nbsp;&nbsp;当为int时，图像上下左右均填充int个，例如padding=4，则上下左右均填充4个pixel，若为3232，则会变成4040。  
&nbsp;&nbsp;&nbsp;&nbsp;当为sequence时，若有2个数，则第一个数表示左右扩充多少，第二个数表示上下的。当有4个数时，则为左，上，右，下。  
fill- (int or tuple) 填充的值是什么（仅当填充模式为constant时有用）。
&nbsp;&nbsp;&nbsp;&nbsp;int时，各通道均填充该值，
&nbsp;&nbsp;&nbsp;&nbsp;当长度为3的tuple时，表示RGB通道需要填充的值。  
padding_mode- 填充模式，这里提供了4种填充模式，  
&nbsp;&nbsp;&nbsp;&nbsp;1.constant（默认），常量。  
&nbsp;&nbsp;&nbsp;&nbsp;2.edge 按照图片边缘的像素值来填充。  
&nbsp;&nbsp;&nbsp;&nbsp;3.reflect，使用图像边缘的镜像反射进行填充。例如，假设图像边缘的像素值是 [1, 2, 3, 4]，则向左填充两个单位会得到 [3, 2, 1, 2, 3, 4]。  
&nbsp;&nbsp;&nbsp;&nbsp;4. symmetric，与 reflect 类似，但填充时不包括最外边的像素值。继续上面的例子，使用 symmetric 模式向左填充两个单位会得到 [2, 1, 1, 2, 3, 4]。  

```python
img01 = transforms.Pad(padding=4)(img96)
img02 = transforms.Pad(padding=8,fill=0)(img96)
img03 = transforms.Pad(padding=16,padding_mode='reflect')(img96)
img04 = transforms.Pad(padding=8,fill=255)(img96)
img05 = transforms.Pad(padding=4,fill=[255,255,0])(img96)

# plt subslot控制按格子显示，前两个参数表示3行两列，第三个参数表示第几个位置，从1开始变换，从左到右，从上到下。
plt.subplot(3,2,1)
plt.imshow(img96)
plt.title("原图")
plt.subplot(3,2,2)
plt.imshow(img01)
plt.title("增加4像素黑边")
plt.subplot(3,2,3)
plt.imshow(img02)
plt.title("增加8像素黑边")
plt.subplot(3,2,4)
plt.imshow(img03)
plt.title("增加16像素反射")
plt.subplot(3,2,5)
plt.imshow(img04)
plt.title("增加8像素白边")
plt.subplot(3,2,6)
plt.imshow(img05)
plt.title("增加4像素黄边")
plt.show()
```

#### 13.修改亮度、对比度和饱和度：transforms.ColorJitter
**class torchvision.transforms.ColorJitter(brightness=0, contrast=0, saturation=0, hue=0)**   
**功能：**  修改修改亮度、对比度和饱和度  
**参数：**  

brightness：亮度调整的幅度。如果是单个数字，亮度会从 [1-brightness,1+brightneww]中随机选择一个系数。也可以通过一个元组 (min,max) 指定亮度调整的范围。
contrast：对比度调整的幅度。使用方式与亮度相同。
saturation：饱和度调整的幅度。使用方式与亮度相同。
hue：色调调整的幅度。不同于上述参数，色调的调整范围是[−0.5,0.5]，通常这个值应该设置得较小。

**class torchvision.transforms.RandomAffine(degrees, translate=None, scale=None, shear=None, resample=False, fillcolor=0)**   
**功能：**  仿射变换  
**参数：**  
degrees：旋转的角度范围。可以指定一个数值，表示图像将在 [−degrees,degrees] 范围内随机旋转；或者指定一个元组 (min,max)，表示旋转角度的具体范围。
translate：平移的比例范围。这是一个元组 (max_horizontal_shift,max_vertical_shift)，每个值应在 [0, 1] 范围内，表示相对于图像宽度和高度的最大平移量。
scale：缩放比例范围。这是一个元组 (min_scale,max_scale)，用于控制图像缩放的程度。
shear：剪切的角度范围。可以是单个数值或元组，用于水平或垂直剪切，或者是 (horizontal_min,horizontal_max,vertical_min,vertical_max)，以允许不同方向的剪切变化。
resample：重采样方法，用于在旋转或缩放后保持图像质量。可选的方法包括 PIL.Image.NEAREST、PIL.Image.BILINEAR 或 PIL.Image.BICUBIC。
fillcolor（或 fill）：在进行变换后可能出现的空白区域的填充颜色。适用于 PIL 图像模式。

```python
img01 = transforms.ColorJitter(
    brightness=0.2,  # 亮度最大变化20%
    contrast=0.3,    # 对比度最大变化30%
    saturation=0.1,  # 饱和度最大变化10%
    hue=0.05         # 色调最大变化5%
)(img96)
img02 = transforms.Grayscale(num_output_channels=1)(img96)
img03 = transforms.RandomAffine (
    degrees=30,                     # 旋转 ±30 度
    translate=(0.1, 0.1),           # 最大平移 10%
    scale=(0.8, 1.2),               # 缩放比例从 0.8 到 1.2
    shear=10                        # 剪切 ±10 度
)(img96)
img04 = transforms.RandomGrayscale(p=0.3)(img96)  # 有 30% 的概率转换为灰度图

# plt subslot控制按格子显示，前两个参数表示3行两列，第三个参数表示第几个位置，从1开始变换，从左到右，从上到下。
plt.subplot(3,2,1)
plt.imshow(img96)
plt.title("原图")
plt.subplot(3,2,2)
plt.imshow(img01)
plt.title("调整亮度对比度饱和度")
plt.subplot(3,2,3)
plt.imshow(img02)
plt.title("灰度变换")
plt.subplot(3,2,4)
plt.imshow(img03)
plt.title("随意变换")
plt.subplot(3,2,5)
plt.imshow(img04)
plt.title("随机灰度")
plt.show()
```
