## MNIST数据集  
MNIST是一个手写数字的数据集，包含了已经标注好的60000张训练集图片和10000张测试集图片，基本类似手写数字识别的Hello World。  

数据集来自美国国家标准与技术研究所, National Institute of Standards and Technology (NIST). 训练集 (training set) 由来自 250 个不同人手写的数字构成, 其中 50% 是高中学生, 50% 来自人口普查局 (the Census Bureau) 的工作人员. 测试集(test set) 也是同样比例的手写数字数据。

MNIST 数据集可在 http://yann.lecun.com/exdb/mnist/ 获取, 它包含了四个部分:

Training set images: train-images-idx3-ubyte.gz (9.9 MB, 解压后 47 MB, 包含 60,000 个样本)
Training set labels: train-labels-idx1-ubyte.gz (29 KB, 解压后 60 KB, 包含 60,000 个标签)
Test set images: t10k-images-idx3-ubyte.gz (1.6 MB, 解压后 7.8 MB, 包含 10,000 个样本)
Test set labels: t10k-labels-idx1-ubyte.gz (5KB, 解压后 10 KB, 包含 10,000 个标签)

在使用pytorch进行学习时，可以使用pytorch的处理图像视频的torchvision工具集直接下载MNIST的训练和测试图片。

MNIST的图像是28x28的灰度图像

### 封装数据集
在 PyTorch 中，**torch.utils.data.DataLoader** 是一个非常重要的类，用于封装数据集，提供对数据的批量处理、打乱顺序和多进程加载等功能。这使得数据加载更加简便高效，特别是在训练大型深度学习模型时，能够有效地提高计算和训练的速度。    

##### 主要功能  
**批量处理：** DataLoader 可以自动将数据分割成批次，每个批次包含一定数量的数据样本。这是训练神经网络时常用的做法，可以在每次迭代中同时处理多个样本。  
**打乱数据：** 在训练过程中，为了使模型训练更有效，通常需要随机打乱输入数据的顺序。DataLoader 提供了简单的选项来实现这一功能。  
**多进程数据加载：** DataLoader 可以利用 Python 的多进程能力，从而异步加载数据，显著提高数据准备的速度，减少等待时间。  
##### 参数说明
**dataset：** 要加载的数据集，必须是继承自 torch.utils.data.Dataset 的类的实例。  
**batch_size：** 每个批次的大小。  
**shuffle：** 布尔值，表示是否在每个 epoch 开始时打乱数据。  
**num_workers：** 用于数据加载的子进程数。多进程可以加速数据加载过程，但也需要谨慎设置以避免成为系统瓶颈。  
**collate_fn：** 如何将多个样本数据拼接成一个批次数据。通常情况下默认的拼接方式已经足够，但对于复杂的数据结构可能需要自定义。  
**drop_last：** 布尔值，表示是否在数据数量不能被批量大小整除时丢弃最后一批数据。 
**pin_memory（可选参数）：** 如果设置为 True，则数据加载到 GPU 时会将数据存储在 CUDA 的锁页内存中，这可以加速数据传输到 GPU。默认值为 False。 
**timeout（可选参数）：** 如果设置为正整数，它定义了每个子进程在等待数据加载器传递数据时的超时时间（以秒为单位）。这可以用于避免子进程卡住的情况。默认值为 0，表示没有超时限制。 
**worker_init_fn（可选参数）：** 一个可选的函数，用于初始化每个子进程的状态。这对于设置每个子进程的随机种子或其他初始化操作很有用。

## 创建神经网络  

```python
class Model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        selectMode = 1
        if selectMode==1:  # 选择第一个模型
            self.conv1 = torch.nn.Sequential(
                torch.nn.Conv2d(1, 10, kernel_size=5),   # 第一层卷积，5x5卷积核，1通道28x28进，10通道24x24出
                torch.nn.ReLU(),                         # 激活函数是ReLU
                torch.nn.MaxPool2d(kernel_size=2),       # 2x2池化层，10通道12x12出
            )
            self.conv2 = torch.nn.Sequential(
                torch.nn.Conv2d(10, 20, kernel_size=5),  # 第二层卷积，5x5卷积核，10通道12x12进，20通道8x8出
                torch.nn.ReLU(),                         # 激活函数是ReLU
                torch.nn.MaxPool2d(kernel_size=2),       # 2x2池化层，20通道4x4出
            )
            self.fc = torch.nn.Sequential(
                torch.nn.Linear(320, 50),                # 全连接层，320通道（20x4x4）入，50通道出
                torch.nn.Linear(50, 10),                 # 全连接层，50通道（20x4x4）入，10通道出
            )
        else:  # 选择第二个模型
            self.conv1 = torch.nn.Sequential(
                torch.nn.Conv2d(1, 32, kernel_size=3),   # 第一层卷积，3x3卷积核，1通道28x28进，32通道26x26出
                torch.nn.ReLU(),                         # 激活函数是ReLU
                torch.nn.MaxPool2d(kernel_size=2),       # 2x2池化层，32通道13x13出
            )
            self.conv2 = torch.nn.Sequential(
                torch.nn.Conv2d(32, 64, kernel_size=3),  # 第二层卷积，3x3卷积核，32通道13x13进，64通道11x11出
                torch.nn.ReLU(),                         # 激活函数是ReLU
                torch.nn.MaxPool2d(kernel_size=2),       # 2x2池化层，64通道5x5出
            )
            self.fc = torch.nn.Sequential(
                torch.nn.Linear(1600, 64),               # 全连接层，1600通道（64x5x5）入，64通道出
                torch.nn.Linear(64, 10),                 # 全连接层，64通道入，10通道出
            )
            
    def forward(self, x):
        batch_size = x.size(0)
        x = self.conv1(x)  # 一层卷积层,一层池化层,一层激活层(图是先卷积后激活再池化，差别不大)
        x = self.conv2(x)  # 再来一次
        x = x.view(batch_size, -1)  # flatten 变成全连接网络需要的输入 (batch, 64,5,5) ==> (batch,1600), -1 此处自动算出的是320
        x = self.fc(x)
        return x  # 最后输出的是维度为10的，也就是（对应数学符号的0~9）
```

### 卷积层  
在 PyTorch 中，**torch.nn.Conv2d** 是一个非常核心的类，用于创建二维卷积层。  
这个层是构建卷积神经网络（CNN）的基础，主要用于处理图像或其他二维数据。卷积层通过在输入数据上滑动（卷积操作）多个过滤器（或称为卷积核），提取特征，这些特征随后可以用于分类、检测等多种计算机视觉任务。  

**参数说明**
torch.nn.Conv2d 的构造函数包含以下主要参数：

**in_channels：** 输入数据的通道数。例如，彩色 RGB 图像的 in_channels 为 3。  
**out_channels：** 卷积产生的通道数，也就是卷积核的数量。每个卷积核都会产生一个不同的特征图。  
**kernel_size：** 卷积核的大小。可以是单一的数字或一个 (height, width) 元组。例如，kernel_size=3 或 kernel_size=(3, 3) 表示使用 3x3 的卷积核。  
**stride：** 卷积时滑动步长。可以是单一数字或 (height, width) 元组。默认值是 1。  
**padding：** 输入数据的周围填充多少层零。可以是单一数字或 (height, width) 元组。默认值是 0。  
**dilation：** 卷积核中元素之间的空间。这对于增大感受野（即卷积核覆盖的区域）很有用。默认值是 1。  

**groups：** 控制输入和输出之间的连接。groups=1 表示所有输入通道都被连接到输出通道，而 groups=in_channels 表示每个输入通道与一个输出通道相连，这称为分组卷积，有助于降低参数数量。  
**bias：** 是否添加可学习的偏置到输出通道。默认为 True。  

在 PyTorch 中，使用 torch.nn.Conv2d 创建的卷积层的卷积核（也称为权重）默认是随机初始化的。这种随机初始化是为了打破对称性并帮助梯度下降算法在训练过程中更有效地找到误差函数的最小值。默认情况下，PyTorch 使用均匀分布或正态分布进行初始化，具体的初始化方式可能会根据 PyTorch 版本和设置有所不同。因为这个权重是在训练过程中不断调整的，所以除非有特殊需要，可以让系统随机初始化。  

**初始化方式**  
卷积层的权重通常通过以下几种方式之一进行初始化：  

均匀分布 (uniform)：权重从一个均匀分布中随机选取，这个分布的范围通常与权重矩阵的大小有关。  
正态分布 (normal)：权重从一个正态分布中随机选取，均值通常是0，标准差与权重矩阵的大小有关。  
**自定义权重初始化**  
如果你希望自定义卷积层的权重初始化，PyTorch 提供了灵活的方式来实现这一点。你可以使用 torch.nn.init 模块中的函数来设置卷积层的权重和偏置。  

自定义初始化可以根据特定的应用需求调整，例如，在某些情况下，特定类型的初始化（如使用 He 初始化或 Xavier 初始化）可能会提高网络的训练速度和性能。这些初始化方法特别适用于深度网络，有助于解决梯度消失或梯度爆炸的问题。  

**如何手动设置权重**  
你可以直接修改卷积层的 weight 属性来指定权重矩阵。例如，如果你想要设置一个用于边缘检测的卷积核，你可以按以下步骤操作：  

创建卷积层：首先定义一个 Conv2d 层。  
设置权重：根据所需的操作手动设置卷积核的值。  
冻结权重（可选）：如果不希望这些权重在训练过程中被修改，可以将它们设置为不可训练（requires_grad=False）。  
例子：

#创建一个 Conv2d 层，1个输入通道，1个输出通道，卷积核大小为3
conv_layer = nn.Conv2d(1, 1, kernel_size=3, padding=1, bias=False)

#定义一个边缘检测的卷积核（垂直边缘检测）
edge_kernel = torch.tensor([
    [-1., 0., 1.],
    [-1., 0., 1.],
    [-1., 0., 1.]
], dtype=torch.float32)

#重新设置卷积核的权重
conv_layer.weight.data[0, 0] = edge_kernel
#将权重的 requires_grad 设置为 False，阻止梯度更新
conv_layer.weight.requires_grad_(False)

这样创建的卷积核权重在训练的时候不会被更新，可以完成对图像的特定操作。

### 激活函数
torch.nn.ReLU是这里选择的激活函数  
#### torch支持的激活函数主要有：（常用的是前3个） 
**ReLU (Rectified Linear Unit)：** 最常用的激活函数之一，定义为 f(x) = max(0, x)。它的主要优点是计算简单和训练过程中收敛速度较快。
torch.nn.ReLU()
**Sigmoid：**  将输入值压缩到 0 和 1 之间，通常用于二分类问题中的输出层。
torch.nn.Sigmoid()
**Tanh (Hyperbolic Tangent)：** 将输入值压缩到 -1 和 1 之间，形状和 Sigmoid 类似，但输出范围更广。
torch.nn.Tanh()
**LeakyReLU：** ReLU 的一个变种，允许在输入小于0时有一个非零的梯度，可以解决 ReLU 的死亡神经元问题。
torch.nn.LeakyReLU(negative_slope=0.01)
**PReLU (Parametric ReLU)：** LeakyReLU 的泛化形式，其负区间斜率是可学习的参数。
torch.nn.PReLU()
**ELU (Exponential Linear Unit)：** 对于负数输入，它使用指数函数减小梯度消失的影响。
torch.nn.ELU(alpha=1.0)
**SELU (Scaled Exponential Linear Unit)：** 自归一化激活函数，输出自动缩放到零均值和单位方差，有助于改善深层网络的训练。
torch.nn.SELU()
**Softmax：** 通常用于多分类问题的输出层，将输入“压缩”成一个概率分布，每个类别的概率都是非负的，并且总和为1。
torch.nn.Softmax(dim=None)，通常需要指定维度 dim。
**Softplus：** 平滑版本的 ReLU，公式为 f(x) = log(1 + exp(x))。
torch.nn.Softplus()
**GELU (Gaussian Error Linear Unit)：** 一种基于高斯分布的激活函数，最近在自然语言处理模型中得到了广泛应用。
torch.nn.GELU()

#### 常用的激活函数优缺点  
**ReLU (Rectified Linear Unit)**  
**优点：**  
计算简单，加速神经网络的训练。  
在正区间内梯度恒定，有助于解决梯度消失问题（在正区间）。  
非饱和性（在正区间）。  
**缺点：**  
死亡ReLU问题（Dead ReLU Problem）：某些神经元可能永远不会激活，导致相应的参数无法更新。  
只能在正区间内激活，输出不是零中心的。  
**适用场景：** 大多数现代深度学习网络中，特别是在不太深的网络中表现良好。  
  
**Sigmoid**  
**优点：**  
输出范围有界（0到1），输出可以解释为概率，适合用于二分类问题的输出层。  
平滑梯度，提供非线性。  
**缺点：**  
强烈的梯度消失问题，由于输出非常接近0或1时梯度几乎为零。
输出不是零中心的。
计算相对较慢。
**适用场景：**  二分类问题的输出层，特别是在网络较浅的情况下。
  
**Tanh (Hyperbolic Tangent)**  
**优点：**  
输出范围有界（-1到1），输出为零中心，这有助于数据的处理和网络的收敛。
平滑梯度，提供非线性。
**缺点：**  
在输入值较大或较小的区域，依然存在梯度消失问题。
**适用场景：** 隐藏层中较为常见，尤其是在要求输出为零中心的场景中较为合适。
  
选择激活函数的一般考虑
任务的性质：例如，使用Sigmoid函数输出概率值适合于二分类任务。
避免梯度消失和爆炸：深层网络更倾向于使用ReLU及其变种，因为它们在正区间内可以缓解梯度消失问题。
数据分布：如果数据（尤其是输入数据）是零中心的，使用Tanh这样的零中心激活函数可能会更好。
计算资源和训练效率：ReLU及其变种通常更快，因为它们计算上不涉及复杂的数学运算。
在实际应用中，选择激活函数常常需要基于实验和实际的网络表现来做决定，不同的问题可能需要不同的激活函数组合。在深度学习的研究和实践中，这仍然是一个活跃的研究领域。

## 损失函数

在 PyTorch 中，损失函数是衡量模型预测值与实际值差异的指标，它是训练过程中优化的关键。根据不同的任务（如分类、回归等），可以选择不同的损失函数。PyTorch 提供了多种内置的损失函数，以下是一些常用的损失函数及其简要说明：  

**1. 均方误差损失（Mean Squared Error Loss, MSE）**  
用途：常用于回归任务。  
函数：torch.nn.MSELoss()  
描述：计算预测值和真实值之间差的平方的均值。  
**2. 交叉熵损失（Cross-Entropy Loss）**  
用途：常用于分类任务。  
函数：torch.nn.CrossEntropyLoss()  
描述：计算目标类别的对数概率和预测概率之间的交叉熵。对于多分类问题非常有效，这个函数内部同时执行了 Softmax 操作。  
**3. 二元交叉熵损失（Binary Cross-Entropy Loss）**  
用途：常用于二分类任务。  
函数：torch.nn.BCELoss()  
描述：计算二分类的交叉熵损失，当使用这个损失函数时，模型的输出应通过 Sigmoid 函数转换为概率。  
**4. 带权重的二元交叉熵损失（Binary Cross-Entropy Loss with Weights）**  
用途：二分类任务，特别是类别不平衡显著时。  
函数：torch.nn.BCEWithLogitsLoss()  
描述：这个损失函数结合了 BCELoss 和 Sigmoid 层。它在计算损失之前先应用 Sigmoid 函数，可以提供更好的数值稳定性。  
**5. 负对数似然损失（Negative Log Likelihood Loss）**  
用途：多分类问题，特别是当模型输出为对数概率（log-probability）时。  
函数：torch.nn.NLLLoss()  
描述：这通常与 LogSoftmax 层一起使用，在这种情况下，模型输出类别的对数概率，然后损失函数计算这些对数概率的负对数似然。  
**6. 平滑 L1 损失（Smooth L1 Loss）**  
用途：回归任务，常用于目标检测中边界框的回归。  
函数：torch.nn.SmoothL1Loss()  
描述：是 L1 损失和 L2 损失的组合，当误差很小的时候表现像 L2 损失，误差大时表现像 L1 损失，减少了对离群点的敏感性。  
**7. 余弦相似度损失（Cosine Similarity Loss）**  
用途：比较两个向量的方向相似度，常用于一些需要度量学习或相似性比较的任务。  
函数：torch.nn.CosineEmbeddingLoss()  
描述：计算两个向量之间的余弦距离损失。  
这些损失函数是优化神经网络时使用的核心组件，选择适合特定任务的损失函数是模型设计的关键步骤。在实际应用中，可能需要根据具体问题调整或自定义损失函数以达到最佳的训练效果。  

## 优化器optimizer
在 PyTorch 中，优化器是用于更新神经网络中的权重和参数以减小损失函数的关键组件。PyTorch 提供了多种优化算法，这些优化器都封装在 torch.optim 模块中。下面是一些常用的优化器及其简要说明：  

**1. SGD (随机梯度下降)**  
类名：torch.optim.SGD  
描述：这是最基本的优化算法，它按照梯度下降的方向更新参数，可选地加入动量（momentum）以加速学习过程。  
主要参数：  
lr (学习率)  
momentum (动量)  
weight_decay (权重衰减，L2正则化)  
nesterov (是否使用 Nesterov 动量)  
**2. Adam (自适应矩估计)**  
类名：torch.optim.Adam  
描述：Adam 是一种自适应学习率的优化算法，结合了 AdaGrad 和 RMSProp 的优点。它通过计算梯度的一阶矩估计和二阶矩估计来调整每个参数的学习率。
主要参数：  
lr (学习率)  
betas (用于计算梯度及其平方的移动指数衰减率)  
eps (数值稳定性的小常数)  
weight_decay (权重衰减)  
**3. AdamW (Adam with Weight Decay)**  
类名：torch.optim.AdamW  
描述：AdamW 是 Adam 的一个变种，它对权重衰减的处理方式与其他参数更新的方式分开，通常在某些情况下能提供更好的训练稳定性和性能。  
主要参数：  
lr (学习率)  
betas (动量因子)  
eps (数值稳定性参数)  
weight_decay (权重衰减)  
**4. RMSprop**    
类名：torch.optim.RMSprop  
描述：RMSprop 是一种自适应学习率方法，特别是设计用来解决 AdaGrad 在处理非凸优化问题时学习率急剧下降的问题。  
主要参数：  
lr (学习率)  
alpha (平滑常数)  
eps (数值稳定性的小常数)  
weight_decay (权重衰减)  
momentum (动量)  
**5. Adagrad**    
类名：torch.optim.Adagrad  
描述：Adagrad 是一种自适应学习率的方法，它可以对低频参数进行大幅更新，对高频参数进行小幅更新，非常适合处理稀疏数据。  
主要参数：  
lr (学习率)  
lr_decay (学习率衰减)  
weight_decay (权重衰减)  
这些优化器的选择取决于具体的应用场景和任务。不同的优化器可能在不同的任务和数据集上表现出不同的性能特点。例如，对于大多数视觉任务，Adam 和 SGD 都是非常流行的选择，而对于一些特定的自然语言处理或稀疏数据任务，Adagrad 和 RMSprop 可能更合适。在实际应用中，经常需要实验比较不同优化器在特定任务上的表现，以选出最佳方案。  

```python
# 训练函数
def train(epoch):
    running_loss = 0.0  # 这整个epoch的loss清零
    running_total = 0
    running_correct = 0
    # enumerate 是一个内置函数，用于将一个可迭代的数据对象（如列表、元组或字符串）组合为一个索引序列，
    # 同时列出数据和数据的下标。这对于同时需要数据和数据索引时非常有用。
    for batch_idx, data in enumerate(train_dl, 0):
        inputs, target = data   # data中包含数据和标签，这里分别取出来
        inputs = inputs.to(device)    # 加载到设备，这里的to（device）要和前面的model一致，否则会出错
        target = target.to(device)    # 加载到设备，这里的to（device）要和前面的model一致，否则会出错
        
        optimizer.zero_grad()     # 这一步把梯度清零

        # forward + backward + update
        outputs = model(inputs)            # 使用训练输入数据计算输出
        loss = criterion(outputs, target)  # 对比已经有的标签target计算损失

        loss.backward()    # 执行反向传播算法的核心。自动计算与当前损失相关的所有参数的梯度。       
        optimizer.step()   # 使用被计算并存储在参数的 .grad 属性中的梯度数据，来更新模型的参数

        # 把运行中的loss累加起来，为了下面300次一除
        running_loss += loss.item()
        # 把运行中的准确率acc算出来
        # torch.max的dim=1指示在每行找最大值，返回最大值和序号，前面我们用_，不关心最大值，把序号保存在predicted里面
        # 如果是分类数据，predicted里面就是第几个分类
        _, predicted = torch.max(outputs.detach(), dim=1) 
        running_total += inputs.shape[0]
        running_correct += (predicted == target).sum().item()

        if batch_idx % 300 == 299:  # 不想要每一次都出loss，浪费时间，选择每300次出一个平均损失,和准确率
            print('[%d, %5d]: loss: %.3f , acc: %.2f %%'
                  % (epoch + 1, batch_idx + 1, running_loss / 300, 100 * running_correct / running_total))
            running_loss = 0.0  # 这小批300的loss清零
            running_total = 0
            running_correct = 0  # 这小批300的acc清零
```

```python
# 测试函数
def test():
    correct = 0
    total = 0
    with torch.no_grad():  # 测试集不用算梯度
        for data in test_dl:
            images, labels = data
            images = images.to(device)   
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, dim=1)  # dim = 1 列是第0个维度，行是第1个维度，沿着行(第1个维度)去找1.最大值和2.最大值的下标
            total += labels.size(0)  # 张量之间的比较运算
            correct += (predicted == labels).sum().item()
    acc = correct / total
    print('[%d / %d]: Accuracy on test set: %.1f %% ' % (epoch+1, EPOCH, 100 * acc))  # 求测试的准确率，正确数/总数
    return acc

```

```python
if __name__ == '__main__':
    acc_list_test = []
    for epoch in range(EPOCH):
        train(epoch)
        # if epoch % 10 == 9:  #每训练10轮 测试1次
        acc_test = test()
        acc_list_test.append(acc_test)

    plt.plot(acc_list_test)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy On TestSet')
    plt.show()
```
